```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import sklearn.metrics 

df= pd.read_csv("Salary Data.csv", sep=',')
df.head()

#EDA
df.isnull().sum()
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary Analysis')
x= df.drop(columns=['Salary'], axis=1)
x.head()
y=df['Salary']
y.head()
#Linear Regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
x.mean()
y.mean()
plt.scatter(x.mean(),y.mean(), color='red')

#Linear Regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
#Train the model
reg.fit(x,y) 
m=reg.coef_
c=reg.intercept_
#y=mx+c
m*1.5+c
reg.predict([[1.5]])
df['Predicted_y']=reg.predict(x)
df.head()
plt.plot(x, reg.predict(x))

#calculate Lose
df['Lose']= df['Salary']-df['Predicted_y']
df.head()

#Cost Calculate(MSE & MAE)
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse= mean_squared_error(df['Salary'], df['Predicted_y'])
mse
mae=mean_absolute_error(df['Salary'], df['Predicted_y'])
mae
plt.scatter(df['Experience Years'], df['Predicted_y'])

#R2 Squared value/Model Performance/ Accuracy for Regression
reg.score(x,y)
```
